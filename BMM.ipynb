```python
import pandas as pd
import numpy as np
```

```python
bd = pd.read_csv('base - с погодой.csv')
```

```python
bd
```

```python
bd['type_of_car'] = bd['type_of_car'].str.replace('ФУСО', 'Фусо')
bd['type_of_car'] = bd['type_of_car'].str.replace('боковая', 'Боковая')
bd['type_of_car'] = bd['type_of_car'].str.replace('бункеровоз', 'Бункеровоз')
bd['type_of_car'] = bd['type_of_car'].str.replace('задняя', 'Задняя')
bd['type_of_car'] = bd['type_of_car'].str.replace('кму', 'КМУ')
bd['type_of_car'] = bd['type_of_car'].str.replace('мультилифт', 'Мультилифт')
bd['type_of_car'] = bd['type_of_car'].str.replace('самосвал', 'Самосвал')
bd['type_of_car'] = bd['type_of_car'].str.replace('фусо', 'Фусо')
bd.head()
```

```python
bd = bd[bd['size_of_trash'] != '0']
bd = bd[bd['mass_of_trash'] != '0']
bd = bd[bd['mass_of_trash'] != '-']
bd['mass_of_trash'] = bd['mass_of_trash'].str.replace(' ', '').str.replace(',', '.').astype(float)
bd['size_of_trash'] = bd['size_of_trash'].str.replace(' ', '').str.replace(',', '.').astype(float)
bd['density'] = bd['density'].str.replace(' ', '').str.replace(',', '.').astype(float)
bd['tmin'] = bd['tmin'].str.replace('°', '').astype(float)
bd['tmax'] = bd['tmax'].str.replace('°', '').astype(float)
bd['taverage'] = (bd['tmax'] + bd['tmin']) / 2
bd.head()
```

```python
bd = bd[bd['density'] < 500]
bd
```
